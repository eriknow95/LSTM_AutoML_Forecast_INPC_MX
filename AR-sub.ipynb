```python
#Modelo AR para pronosticar el INPC Subyacente
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.tsa.ar_model import AutoReg
from pmdarima.arima import auto_arima
import matplotlib.pyplot as plt
#Todos los tÃ­tulos en times new roman
plt.rc('font', family='Times New Roman')
```

```python
# Leer los datos
df = pd.read_excel('IPC.xlsx', sheet_name='IPCSUB', index_col= 'fecha',parse_dates=['fecha'])


```

```python
# Verificar la estacionariedad de la serie
def check_stationarity(df):
    result = sm.tsa.adfuller(df)
    print('ADF Statistic: {:.3f}'.format(result[0]))
    print('p-value: {:.3f}'.format(result[1]))
    print('Critical Values:')
    for key, value in result[4].items():
        print('\t{}: {:.3f}'.format(key, value))
        
check_stationarity(df['ipcsub'])

```

```python
# Seleccionar el orden del modelo AR utilizando AIC y BIC
order_aic = sm.tsa.arma_order_select_ic(df['ipcsub'], ic='aic')['aic_min_order']
order_bic = sm.tsa.arma_order_select_ic(df['ipcsub'], ic='bic')['bic_min_order']

print("Orden del modelo AR (AIC):", order_aic)
print("Orden del modelo AR (BIC):", order_bic)

```

```python
# Ajustar el modelo AR seleccionado
modelo_ar = AutoReg(df['ipcsub'][:'2021'],lags = [3])
modelo_ar_entrenado = modelo_ar.fit()  # Usamos el orden AIC para ajustar el modelo

```

```python
print(modelo_ar_entrenado.summary())
```

```python
start_date = pd.to_datetime('2022-01-31')
end_date = pd.to_datetime('2022-12-31')
forecast = modelo_ar_entrenado.predict(start=start_date, end=end_date)
```

```python
forecast=  pd.DataFrame({'fecha': forecast.index, 'ipcsub': forecast.values})
forecast.set_index(['fecha'], inplace=True)
```

```python
from sklearn.metrics import mean_squared_error
test_data=pd.DataFrame(df['ipcsub']['2022-01':'2022-12'])
mse = mean_squared_error(test_data, forecast)
print('El MSE de la predicciÃ³n es:', mse)
```

```python
plt.figure(figsize=(10, 5))

item_id = "ipcsub"
y_past = df[item_id]
y_pred = forecast[item_id]['2022':]
y_test = test_data[item_id]['2022-01':'2022-12']

plt.plot(y_past['2020-10':'2022'], label="Valores pasados")
plt.plot(y_pred, label="PredicciÃ³n")
plt.plot(y_test, label="Valores futuros")
plt.title('INPC Subyacente / PredicciÃ³n 2022/ Modelo AR')

plt.legend();
```

```python
# Seleccionar el orden del modelo AR utilizando AIC y BIC
order_aic = sm.tsa.arma_order_select_ic(df['ipcsub'][:'2007'], ic='aic')['aic_min_order']
order_bic = sm.tsa.arma_order_select_ic(df['ipcsub'][:'2007'], ic='bic')['bic_min_order']

print("Orden del modelo AR (AIC):", order_aic)
print("Orden del modelo AR (BIC):", order_bic)

```
